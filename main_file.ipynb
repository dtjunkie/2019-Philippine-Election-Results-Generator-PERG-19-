```python
import pandas as pd
```

```python
def region_dir():
    df = pd.read_json('comelec-2019-election-results/data/regions/root.json', lines=True,
                 nrows=1)
    regions = pd.DataFrame(df['srs'][0]).T
    regions = dict(zip(regions['rn'],regions['rc']))
    return regions
```

```python
print(region_dir())
```

```python
def province_dir(region):
    region_code = region_dir().get(region)
    df = pd.read_json(f'comelec-2019-election-results/data/regions/0/{region_code}.json', lines=True, 
                             nrows=1)
    provinces = pd.DataFrame(df['srs'][0]).T
    provinces = dict(zip(provinces['rn'],provinces['rc']))
    return provinces
```

```python
print(province_dir('OAV'))
```

```python
def citymun_dir(region, province):
    province_code = province_dir(region).get(province)
    df = pd.read_json(f'comelec-2019-election-results/data/regions/0/{province_code}.json', lines=True, 
                             nrows=1)
    citymuns = pd.DataFrame(df['srs'][0]).T
    citymuns = dict(zip(citymuns['rn'],citymuns['rc']))
    return citymuns
```

```python
print(citymun_dir('REGION I','ILOCOS NORTE'))
```

```python
def barangay_dir(region, province, citymun):
    citymun_code = citymun_dir(region, province).get(citymun)
    
    try:
        df = pd.read_json(f'comelec-2019-election-results/data/regions/0/{citymun_code}.json', lines=True, 
                          nrows=1)
    except:
        df = pd.read_json(f'comelec-2019-election-results/data/regions/1/{citymun_code}.json', lines=True, 
                                 nrows=1)
        
    barangays = pd.DataFrame(df['srs'][0]).T
    barangays = dict(zip(barangays['rn'],barangays['rc']))
    return barangays
```

```python
print(barangay_dir('REGION XIII','AGUSAN DEL SUR','BAYUGAN  CITY'))
```

```python
def precinct_dir(region, province, citymun, barangay):
    brgy_code = barangay_dir(region, province, citymun).get(barangay)
    folder_number = 1
    try:
        df = pd.read_json(f'comelec-2019-election-results/data/regions/{folder_number}/{brgy_code}.json', 
                          lines=True, nrows=1)
    except:
        folder_number += 1
        while folder_number < 44:
            try:
                df = pd.read_json(f'comelec-2019-election-results/data/regions/{folder_number}/{brgy_code}.json', 
                          lines=True, nrows=1)
            except:
                folder_number += 1
            else:
                break
                
    precincts = pd.DataFrame(df['pps'][0])
    precincts = precincts.sort_values(by='ppc',ascending=True)
    precincts = dict(zip(precincts['ppc'],precincts['ppn']))
    return precincts
```

```python
def senatorial_candidates_list():
    df = pd.read_json(f'comelec-2019-election-results/data/contests/1.json')
    sen_candidates = pd.DataFrame(list(df['bos']))
    sc_split = sen_candidates['bon'].str.split(pat="(", expand=True)
    sen_candidates['candidate_name'] = sc_split[0]
    sen_candidates['party_abbr'] = [party.replace(")","") for party in sc_split[1]]
    sen_candidates['contest_code'] = df['cc']
    sen_candidates['contest_desc'] = df['cn']
    sen_candidates = sen_candidates[['contest_code','contest_desc','boc','candidate_name','pc','party_abbr','pn']]
    sen_candidates = sen_candidates.sort_values(by='boc', ascending = True)
    sen_candidates = sen_candidates.rename(columns={'boc':'candidate_code','pc':'party_code','pn':'party_name'})
    sen_candidates = sen_candidates.reset_index(drop=True)
    return sen_candidates
```

```python
def partylist_rep_list():
    df = pd.read_json('comelec-2019-election-results/data/contests/5567.json')
    pl_candidates = pd.DataFrame(list(df['bos']))
    plc_split = pl_candidates['bon'].str.split(pat=" ", n=1, expand=True)
    pl_candidates['pl_number'] = plc_split[0]
    pl_candidates['party_name'] = plc_split[1]
    pl_candidates['contest_code'] = df['cc']
    pl_candidates['contest_desc'] = df['cn']
    pl_candidates = pl_candidates[['contest_code','contest_desc','boc','pl_number','party_name']]
    pl_candidates = pl_candidates.sort_values(by='boc', ascending=True)
    pl_candidates = pl_candidates.rename(columns={'boc':'candidate_code'})
    pl_candidates = pl_candidates.reset_index(drop=True)
    return pl_candidates
```

```python
def local_candidates_list(region, province, citymun):
    citymun_code = citymun_dir(region, province).get(citymun)
    
    try:
        df = pd.read_json(f'comelec-2019-election-results/data/regions/0/{citymun_code}.json', lines=True, 
                          nrows=1)
    except:
        df = pd.read_json(f'comelec-2019-election-results/data/regions/1/{citymun_code}.json', lines=True, 
                                 nrows=1)
        
    contest_codes = df['cs'][0]
    data_list = []
    
    for code in contest_codes:
        if code != 1 and code !=5567:
            df2 = pd.read_json(f'comelec-2019-election-results/data/contests/{code}.json')
            data_list.append(df2)
            df3 = pd.concat(data_list)
            
    df3 = df3.sort_values(by = 'cc', ascending = True).reset_index(drop=True)
    loc_candidates = pd.DataFrame(list(df3['bos']))
    lc_split = loc_candidates['bon'].str.split(pat="(", expand=True)
    loc_candidates['candidate_name'] = lc_split[0]
    loc_candidates['party_abbr'] = [party.replace(")","") for party in lc_split[1]]
    loc_candidates['contest_code'] = df3['cc']
    loc_candidates['contest_desc'] = df3['cn']
    loc_candidates = loc_candidates.sort_values(by='boc', ascending = True)
    loc_candidates = loc_candidates[['contest_code','contest_desc','boc','candidate_name','pc','party_abbr','pn']]
    loc_candidates = loc_candidates.rename(columns={'boc':'candidate_code','pc':'party_code','pn':'party_name'})
    
    return loc_candidates
            
```

```python
def precinct_results(contest_code, region, province, citymun, barangay,pct=0):
    data_list = []
    for code in precinct_dir(region, province, citymun, barangay).keys():
        if 1 <= code <= 999:
            df = pd.read_json(f'comelec-2019-election-results/data/results/0/{code}.json', lines=True, nrows=1)
        elif 1000 <= code <= 9999:
            folder_number = str(code)[0]
            df = pd.read_json(f'comelec-2019-election-results/data/results/{folder_number}/{code}.json', lines=True, nrows=1)
        else:
            folder_number = str(code)[:2]
            df = pd.read_json(f'comelec-2019-election-results/data/results/{folder_number}/{code}.json', lines=True, nrows=1)
        
        
        df = pd.DataFrame(df['rs'][0])
        df = df[df['cc']==contest_code]
        
        if pct == 0:
            df = df[['v']].T
        elif pct == 1:
            df = df[['per']].T
        
        df['vbc'] = code
        df = df.set_index('vbc',drop=True)
        data_list.append(df)
        result = pd.concat(data_list)
        
        if contest_code == 1:
            name_split = senatorial_candidates_list()['candidate_name'].str.split(pat=",",expand=True)
            name_col = [name.lower().replace(" ","_") for name in name_split[0]]
        elif 1 < contest_code < 5567:
            df2 = local_candidates_list(region,province,citymun)
            df2 =df2[df2['contest_code']==contest_code]
            name_split = df2['candidate_name'].str.split(pat=",",expand=True)
            name_col = [name.lower().replace(" ","_") for name in name_split[0]]
        else:
            name_col = partylist_rep_list()['party_name']
            
        
        result = result.rename(columns=dict(zip(result.columns,name_col)))
        
    return result
```

```python
precinct_results(5567,'REGION XIII','AGUSAN DEL SUR','BAYUGAN  CITY', 'POBLACION')
```

```python
def barangay_results(contest_code,region,province,citymun,barangay,pct=0):
    df = precinct_results(contest_code, region, province, citymun, barangay,pct=0)
    df.loc[barangay] = pd.Series(df.sum(), index = list(df.columns))
    barangay_result = pd.DataFrame(df.loc[barangay]).T
    return barangay_result
```
